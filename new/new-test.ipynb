```python
clu = CompositeLearnUnit(
    llm,
    main_goal="Find the hidden rule and learn the transformation logic of the cryptographic puzzle",
    name="PuzzleSolverCLU3"
)

# Perform training
query = """Given:
"Neural networks transform data efficiently" → "eeraf"
"Artificial intelligence automates decisions" → "rnue" 
Query:
What is "Gradient descent optimizes loss functions" → ?"""

expected_output = "repou"

training_result = clu.train(query, expected_output, OperationalAgentOutput, verbose=True)
```

```python
# Perform reasoning
from pydantic import BaseModel, Field
class OperationalAgentOutput(BaseModel):
    response: str = Field(
        ..., description="The response generated by the Operational Agent"
    )
result=clu.reason("What is the transformation rule for the word puzzle? answer in a concise manner",schema=OperationalAgentOutput,verbose=True)
print(result)
```
